### 컴퓨터 비전에서 인공지능을 활용하는 분야
1. Object Classification : 이미지 속 객체를 인지하여 그 클래스를 분류해내는 기술로 컴퓨터 비전영역에서 가장 기초적인 응용 분야
2. Object Detection & Localization :이미지 또는 비디오 영상에서 객체를 식별해내는 기술

4. Image Captioning: : 이미지의 상황을 텍스트로 설명할 수 있는 기술
5. Object Tracking : 비디오 영상 내의 객체의 위치 변화를 추적하는 기술로 흔
히 포인트 추적(point tracking), 커널 추적(kernel tracking), 실루엣 추적 (silhouette tracking) 등의 방법을 사용
6. Action Classification : 비디오 영상 내의 객체의 행동(action)을 인식하여 분
류하는 기술

## 이미지 읽어오기
` cv2.imread(filename[,flags]) => retval `
- 이미지를 메모리로 읽어오는 함수로, filename은 읽어올 이미지 파일 경로
- flags는 옵션으로, 다음과 같이 설정이 가능합니다.
    - cv2.IMREAD_COLOR(컬러)
    - cv2.IMREAD_GRAYSCALE(흑백)
    - cv2.IMREAD_UNCHANGED(알파 채널 포함)
- 읽어내면 numpy의 ndarray타입으로 결과를 리턴합니다.

## 이미지 저장하기
`cv2.imwrite(filename,img[,params]) => retval`
- 흑백으로 읽어내면 0-255 사이의 값으로 각 픽셀을 표현
- 컬러로 읽어내면 BGR 순서 대로 값을 리턴합니다.
- RGB형태로 변경하고자 할 때는 `cv2.cvtColor`함수를 이용합니다.

## 이미지를 출력하는 방법
```
1. matplotlib.pyplot.imshow(numpy.ndarray)
2. cv2.imwrite(filename,img[,params]) => retval : 파일에 기록
3. cv2.namedWindow(winname[,flags]) => retval : 윈도우 생성
4. cv2.imshow(winname, mat) : 윈도우 이름을 주면 윈도우에 이미지 출력
5. cv2.watitKey([,delay]) =>retval : 키보드 입력 대기
6. cv2.destroyWindow(winname) : 윈도우 파괴
7. cv2.destroyAllWindows() : 모든 윈도우 파괴
```

````python

import cv2
import numpy as np

# 그래픽을 출력할 배경 생성
# numpy의 ndarray로 생성
# zeros는 검정색 화면. 255를 더해서 흰색 화면으로 만들어줌
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8) + 255
type(img) # ndarray
img.shape

#  사각형 좌표
x1, x2 = 100, 400
y1, y2 = 100, 400

# 사각형 그리기
# 이미지, xy좌표, BGR컬러
rect_arr = cv2.rectangle(img, (x1, y1), (x2, y2),(0,255,0))
rect_arr.shape

# 사각형 그릴 좌표 생성 
pt1 = 120, 50   # 첫번째 좌표
pt2 = 300, 500   # 두번째 좌표

# 선 그릴 좌표 생성
# 이미지, 선 좌표1, 선 좌표2, BGR컬러, 두께
cv2.line(img, pt1, pt2, (255, 0, 0), 2)
imgRect = (x1, y1, x2-x1, y2-y1)
# 사각형과 직선의 교차점 계산
# 교차하면  retval:True, 교차점xy:rpt1과 rpt2
retval, rp1, rp2 = cv2.clipLine(imgRect, pt1, pt2)

# 교차하면 교차점 그리기
if retval :
    # 이미지, 좌표, 반지름, GBR컬러, 두께
    cv2.circle(img, rp1, radius=5, color=(0, 255, 0), thickness=-1)
    cv2.circle(img, rp2, radius=5, color=(0, 255, 0), thickness=-1)
# 그리기
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
````

### 타원 그리기

````python
# 그래픽을 출력할 배경 생성
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8) + 255


# 타원 그릴 좌표 생성
# 중앙점 좌표 설정
cx = img.shape[0]// 2    # 무한소수 안 들어가게 주의!
cy = img.shape[1]// 2


# 일반 원생성
# 이미지, 좌표, 반지름, GBR컬러, 두께
cv2.circle(img, (cx, cy), 100, color=(255, 0, 0), thickness=1)
# 타원 객체 생성
# 이미지, 타원의 중심, 축 절반 크기, 각도, 시작 각도, 끝 각도, GBR컬러
cv2.ellipse(img, (cx, cy), (100, 50), 45, 0, 360, (255, 0, 255))
cv2.ellipse(img, (cx, cy), (50, 100), 0, 0, 360, (0, 0, 255))

# 호 객체 생성
# ellipse2Poly(center, axes, angle, arcStart, arcEnd, delta) -> pts
cv2.ellipse2Poly((cx, cy), (0, 50), 30, 0, 30, 10)

# 그리기
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
````

### 다각형 그리기


````python
# 그래픽을 출력할 배경 생성
img = np.zeros(shape=(512, 512, 3), dtype=np.uint8) + 255
# 다각형 그릴 좌표 생성
x, y = 256, 256
size = 200

for angle in range(0, 90, 5):
    rect = ((256, 256), (size, size), angle)
    box = cv2.boxPoints(rect).astype(np.int32)

    r = np.random.randint(256)
    g = np.random.randint(256)
    b = np.random.randint(256)

    # 다각체 객체 생성
    # 이미지, 좌표, isClosed, GRB컬러, 두께
    cv2.polylines(img, [box], True, (r, g, b), 2)
# 그리기
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
````

### 키보드 이벤트 처리

1)Event: 사용자 또는 시스템이 발생시키는 행위

2)Event Handling: 이벤트가 발생했을 때 호출되는 코드를 작성하는 것

3)CallBack Method: 이벤트가 발생하면 호출되는 코드의 모임(메소드)

4)Event Handler 또는 Listener: 이벤트가 발생하면 그 이벤트를 처리하기 위한 코드를 가진 객체

5)Sender: 이벤트가 발생한 객체

6)Event Bubbling: 포함하고 있는 객체 와 포함된 객체에 동일한 이벤트를 처리하기 위한 코드가 존재하는 경우 포함된 객체에 이벤트가 발생하면 자신의 이벤트를 처리하기 위한 코드 와 자신을 포함하고 있는 객체의 이벤트를 처리하기 위한 코드가 연속적으로 호출되는 것

7)Default Event: 시스템이나 프레임워크가 설정해둔 이벤트 처리 코드
web page 에서 submit 버튼을 만들면 submit 버튼을 누르면 form 의 데이터가 전송됩니다.


### Open CV 에서의 이벤트
1)Keyboard 이벤트
cv2.waitKey(): 키보드 입력을 대기하고 키보드를 누르면 1byte 키 값을 리턴합니다. - 누른 키를 비교할 때 정수 값과 비교하던가 문자를 ord 함수로 감싸서 비교해야 합니다.

cv2.waitKeyEx(): 키보드 입력을 대기하고 키보드를 누르면 2byte 키 값을 리턴합니다.



````python
#i 키를 누르면 이미지를 반전하기
import cv2
import numpy as np

#키보드 이벤트 처리할 때 주의할 점은 대소문자 구별 문제

img = cv2.imread('../data/images/lena.jpg', cv2.IMREAD_GRAYSCALE)

cv2.imshow('image', img)

while True:
    #키보드 입력받기
    # 1byte의 입력을 받습니다
    keycode = cv2.waitKey()
    print(keycode)
    if keycode == ord('i') or keycode == ord('I'):
        img = ~img
        cv2.imshow('image', img)
    elif keycode == 27:
        break

cv2.destroyAllWindows()
````

## Mouse Event 처리
- cv2.setMouseCallback(windowName, onMouse[, param])
    - windowName 은 마우스 이벤트를 처리할 윈도우 이름
    - onMouse 는 호출될 함수로 첫번째 매개변수를 마우스 이벤트 이름이고 두번째 와 세번째 매개변수는 마우스 좌표이고 네번째는 flag 로 같이 누른 키값인데 cv2.EVENT_FLAG_SHIFTKEY 또는 CTRLKEY 등입니다.
- 다섯번째 매개변수가 callback을 등록할 때 [, param] 에 설정한 값으로 list 형태로 넘어옵니다.
    - 마우스 이벤트 이름은 cv2.EVENT_LBUTTONDOWN 과 같은 형태입니다.
