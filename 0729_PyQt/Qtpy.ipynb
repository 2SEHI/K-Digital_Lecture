##### setNativeMenuBar(False)
- macOS에서도 Windows 환경과 동일한 결과를 얻음


```python
from PyQt5.QtWidgets import QMainwindow, QAction, qApp
from PyQt5.QtGui import QIcon

# 윈도우를 만들기 위한 클래스
# QWidget는 창만 만들어지고 
# QMainwindow는 창안에 영역이 생깁니다.
class MyApp(QWidget):
    #생성자에서 UI를 생성하고 이벤트 연결
    def __init__(self):
        super().__init__()
        self.setWindowTitle("윈도우 화면 구성")
        
        # menubar 생성 _ ㅁ x 가 있는 영역
        menubar = self.menubar()
        # macOS에서도 Windows 환경과 동일한 결과를 얻음
        menubar.setNativeMenuBar(False)
        # filemenu 생성
        filemenu = menubar.addMenu('&File')

        # 맨 하단의 상태 표시줄
        self.statusBar().showMessage('준비')
        
        # 동작
        # 아이콘 가져오기
        exitAction = QAction(QIcon('/data/exit.png'),'Exit','self')
        # shorcut지정
        exitAction.setShortcut('Ctrl + Q')
        # 
        exitAction.setStatusTip('프로그램 종료')
        # 이 동작을 선택했을 때, 생성된 (triggered) 시그널이 QApplication 위젯의 quit() 메서드에 연결되고, 어플리케이션을 종료시키게 됩니다.
        exitAction.triggered.connect(qApp.quit)
        
        
        self.setGeometry(300, 300, 400, 200)
        self.show() #화면에 출력
        
if __name__ == '__main__':
    app = QtCore.QCoreApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```

##### QLabel
- 텍스트 또는 이미지 라벨을 만들 때 쓰입니다. 

```python
import sys

# 윈도우를 만들기 위한 클래스
# QWidget는 창만 만들어지고 
# QMainwindow는 창안에 영역이 생깁니다.
class MyApp(QWidget):
    #생성자에서 UI를 생성하고 이벤트 연결
    def __init__(self):
        super().__init__()
        self.setWindowTitle("윈도우 화면 구성")
        lbl_red = QLabel('빨강')
        # 영역에 스타일 주기
        lbl_red.setStyleSheet('color:red;',
                             'border-style:solid;',
                             'border-width:2px;',
                             'border-color:#FA8072;',
                             'border-radius:3px')
        lbl_blue = QLabel('녹색')
        lbl_green = QLabel('파랑')
        
        # 세로방향으로 배치
        vbox = QVBoxLayout()
        vbox.addWidget(lbl_red)
        vbox.addWidget(lbl_green)
        vbox.addWidget(lbl_blue)
        self.setLayout(vbox)
        self.setWindowTitle('Stylesheet')
        self.setGeometry(300, 300, 400, 200)
        self.show() #화면에 출력
        
if __name__ == '__main__':
    app = QtCore.QCoreApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())

    
```

```python
import sys
from PyQt5.QtWidgets import * 
from PyQt5.QtCore import * 
from PyQt5.QtGui import * 
from PyQt5.QtWidgets import * 
from PyQt5.QtWidgets import * 
from PyQt5 import uic
from PyQt5.QtWidgets import QMainWindow

import urllib.request
# 앞에서 만든 UI파일을 저장
form_class = uic.loadUiType('main.ui')[0]
print(form_class)
class MyApp(QMainWindow, form_class):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        self.btn_loadFromFile.clicked.connect(self.loadImageFromFile)
        self.btn_loadFromFile.clicked.connect(self.loadImageFromWeb)
    def loadImageFromFile(self):
        self.qPixmapIFileVar = QPixmap()
        self.qPixmapIFileVar.load('hero.jpeg')
        self.setPixmap(self.qPixmapIFileVar)

    def loadImageFromWeb(self):
        urlString = 'http://idchowto.com/wp-content/uploads/2015/11/1234.png'
        imageFromWeb = urllib.request.urlopen(urlString).read()

        self.qPixmapIFileVar = QPixmap()
        self.qPixmapIFileVar.loadFromData(imageFromWeb)
        self.setPixmap(self.qPixmapIFileVar)    
if __name__ == '__main__':
#     app = QtCore.QCoreApplication.instance()
#     if app is None:
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```

## GUI 프로그래밍 방식
- 화면 디자인과 동적이 코드를 프로그래밍 언어의 코드로만 작성
- 디자인은 어렵지만 코드의 가독성은 높아집니다.
- 화면 디자인을 별도의 파일에 하고 동적인 코드를 다른 파일에 작성하는데 디자인은 쉽지만 코드의 가독성이 떨어집니다.
- 디자인 파일은 XMl형식으로 만들어진 경우가 많습니다.
- XML을 직접 수정해서 디자인을 변경할 수 있습니다.
- 실행할 때나 처음 컴파일할 떄 내용을 읽어서 프로그래밍 언어 코드로 변경해서 동적인 코드에 삽합하는 구조로 실행됩니다.
